```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import math
matplotlib.rcParams['font.family'] = 'AppleGothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
def main():
    # 프로그램 기본 설명
    print("이것은 계단 전위함수 입니다.")
    print("전자의 운동에너지, x=0에서의 전위, 그래프의 범위(x = -2nm ~ 사용자 설정)가 정해지면 이에 맞는 그래프가 그려집니다.")
    print("그래프가 그려진 후에는 x = 0 이후의(전위가 있을 때) 원하는 위치에서의 전자 발견 확률을 구할 수 있습니다.")
    print("나노미터를 나타내기 위해서는 x1e-x2의 형태를 넣습니다. 예) 2나노미터 : 2e-9, 0.4나노미터 : 4e-10")
    print("입력 예시: 운동에너지 : 2, 전위 장벽 : 5, 그래프 끝단 : 2e-9, 전자 존재 확률을 확인할 위치 : 2e-10")
    
    kinetic = float(input('운동에너지(eV):'))  # 전자의 운동에너지 입력
    mass = 9.1e-31                          # 전자의 질량 (kg)
    barrier = float(input('전위 장벽(eV):'))  # 전위 장벽
    planck = 1.054e-34                      # 플랑크 상수
    end = float(input('그래프 끝단(나노미터로 예시 : 2nm(2e-9)):'))           # 그래프 끝단 입력(nm)
    x1 = np.linspace(-(2e-9), 0, 1000)      #  x 범위 : -2 나노미터 ~ 0까지 1/1000 단위로 나열
    x2 = np.linspace(0, end, 1000)          #  x 범위 : 0 ~ 입력한 그래프 끝단까지 1/1000 단위로 나열
    k1 = math.sqrt(2*mass*kinetic*(1.6e-19)/math.pow(planck,2))            # 자유공간에서의 파수
    k2 = math.sqrt(2*mass*(barrier-kinetic)*(1.6e-19)/math.pow(planck,2))  # 전위 내부의 파수
    WaveFunction1 = (-1)*(k2/k1)*np.sin(k1*x1) + np.cos(k1*x1)             # 자유공간에서의 파동함수
    WaveFunction2 = np.exp(-(k2*x2))                                       # 전위 내부의 파동함수
    x = float(input('전자 존재 확률을 확인할 위치, x:'))     # 확인하고자 하는 전자 발견 확률의 위치(nm)
    print("해당 위치에서 전자가 발견될 확률은",round(np.exp(-(k2*x))*100, 4),"% 입니다.")    # 전자 발견 확률 출력
    plt.figure(figsize=(15,5))                        # 그래프 크기 조절 (15x5)
    plt.plot(x1, WaveFunction1, color='r')            # 자유공간에서의 파동함수 그래프
    plt.plot(x2, WaveFunction2, color='r')            # 전위 내부에서의 파동함수 그래프
    plt.title('계단 전위 함수')
    plt.xlabel('위치(nm)')
    plt. ylabel('파동함수')
    plt.show()      # 그래프 시각화
    
main()
```

```python
def main():
    # 프로그램 기본 설명
    print("이것은 tunneling effect 입니다.")
    print("전자의 운동에너지, x=0 에서의 전위, 전위 장벽의 길이가 정해지면 그에 맞는 파동함수 그래프가 그려지며, \
    전위 장벽을 통과할 확률을 알려줍니다.")
    print("가로축 x는 나노미터 단위이며, 값을 입력할 때는 '숫자e-9' 로 입력하면 됩니다.")
    print("전자의 운동에너지와 전위의 단위는 eV 단위입니다.")
    
    kinetic = float(input('운동에너지(eV):'))   # 전자의 운동에너지 입력
    barrier = float(input('전위 장벽(eV):'))   # 전위 장벽
    mass = 9.1e-31                           # 전자의 질량 (kg)
    planck = 1.054e-34                       # 플랑크 상수
    width = float(input('전위 장벽 길이(나노미터로 예시 : 0.5nm(5e-10)):'))   # 전위 장벽 길이(nm)
    x1 = np.linspace(-(2e-9), 0, 1000)       #  x 범위 : -2 ~ 0까지 1/1000 단위로 나열 (Region 1)
    x2 = np.linspace(0, width, 1000)         #  x 범위 : 0 ~ 전위 장벽 길이까지 1/1000 단위로 나열 (Region 2)
    x3 = np.linspace(width, 3e-9, 1000)      #  x 범위 : 전위 장벽 길이 ~ 3nm까지 1/1000 단위로 나열 (Region 3)
    k1 = math.sqrt(2*mass*kinetic*(1.6e-19)/math.pow(planck,2))               # 자유공간에서의 파수
    k2 = math.sqrt(2*mass*(barrier-kinetic)*(1.6e-19)/math.pow(planck,2))     # 전위 내부의 파수
    
    # 전위 장벽 통과 후의 파동함수의 계수는 다소 복잡한 수식으로 이루어져 있습니다.
    # 복잡한 수식을 complex_coefficient 이름으로 풀어냈으며 이에 해당하는 부분은 수식 풀이에서 확인할 수 있습니다.
    complex_coefficient = (k1*np.sin(k1*width)-k2*np.cos(k1*width))/(k1*np.cos(k1*width)+k2*np.sin(k1*width))
    
    # F와 G는 Region 3에서 파동함수의 계수입니다. 자세한 수식은 수식 풀이에서 확인할 수 있습니다.
    F = complex_coefficient*(np.exp(-k2*width)/(complex_coefficient*np.sin(k1*width) + np.cos(k1*width)))
    G = np.exp(-k2*width) / (complex_coefficient*np.sin(k1*width) + np.cos(k1*width))
    WaveFunction1 = (-0.3)*(k2/k1)*np.sin(k1*x1) + 0.3*np.cos(k1*x1)       # 자유공간(Region 1)에서의 파동함수
    WaveFunction2 = 0.3*np.exp(-(k2*x2))                                   # 전위 장벽(Region 2)에서의 파동함수
    Wavefunction3 = 0.3*F*np.sin(k1*x3) + 0.3*G*np.cos(k1*x3)              # 자유공간(Region 3)에서의 파동함수
    plt.figure(figsize=(15,5))                                             # 그래프 크기 (15x5)
    plt.plot(x1, WaveFunction1, color='r')                                 # 자유공간(Region 1)에서의 파동함수 그래프
    plt.plot(x2, WaveFunction2, color='r')                                 # 전위 장벽(Region 2)에서의 파동함수 그래프
    plt.plot(x3, Wavefunction3, color='r')                                 # 자유공간(Region 3)에서의 파동함수 그래프
    plt.plot([0, 0], [-0.4, 0.4], color='b', ls='--')                      # 전위 장벽 start point
    plt.plot([width, width], [-0.4, 0.4], color='b', ls='--')              # 전위 장벽 end point
    plt.title('Tunneling effect')
    plt.xlabel('위치(nm)')
    plt.ylabel('파동함수')
    plt.show()
    print("전위장벽을 통과할 확률은", round(np.exp(-(k2*width))*100, 3), "% 입니다.")    # 전위 장벽 통과 확률 출력
    
main()
```
